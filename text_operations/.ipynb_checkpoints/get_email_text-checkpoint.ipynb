```python
def parseOutAllText(f):
    f.seek(0)  ### go back to beginning of file (annoying)
    all_text = f.read()
    ### split off metadata
    content = all_text.split("X-FileName:")
    if len(content) > 1:
        text_string = content[1].translate(string.maketrans("", ""), string.punctuation)
    return text_string
```

```python
# load up detective values, process, and get word usages...
with open("/home/allan/temp_stuff.pkl","r") as f:
    (words,values,from_people_list,to_people_list)=pickle.load(f)

detective_people_list = to_people_list+from_people_list
usage = [0]*len(words)
for person in list(set(detective_people_list)):
    p_index= np.where(np.array(np.array(detective_people_list) == person))[0]
    p_values = np.zeros(len(words))
    #print 'p_index: ',p_index
    for ind in p_index:
        #print "ind",ind
        p_values += values.getrow(ind)
    usage += p_values > 0
```

```python
# SET PARAMETERS
max_usage = 10
min_usage = 2

keep_index_v1 = ((np.array(usage) < max_usage) & (np.array(usage) > min_usage))[0]
keep_index_v2 = np.array(((np.array(usage) < max_usage) & (np.array(usage) > min_usage)).nonzero()[0])
keep_values = values[:,keep_index_v2]
keep_words = np.array(words)[keep_index_v1]
```

```python
detective = {"words":keep_words}
for person in list(set(detective_people_list)):
    person_index = np.array(np.array(np.array(np.array(detective_people_list)==person)).nonzero()[0])[0]
    detective[person] = keep_values[person_index]
with open("/home/allan/Desktop/enron_fraud_detection/data/detective.pkl","w") as f:
    pickle.dump(detective,f)
```

```python
### define qualitative filter here...
from nltk import *
from nltk.stem.snowball import SnowballStemmer

stemmer = SnowballStemmer("english")

def remove_bs(word_list):
    return(word_list)

def qualitative_filter(text):
        word_list=text.split()
        word_list=remove_bs(word_list)
        #filtered_words = [word for word in word_list if word not in stopwords.words('english')]
        cleaned_text=""
        for word in word_list:
            cleaned_word = stemmer.stem(word)
            cleaned_text+=" "+cleaned_word
        return cleaned_text
```

```python
### get "qualitative" "bag of words" representation of the email text.
### In this mode, we are concerned only with the relatively common 
### vocabulary words in the text written by the individual, and the
### relative frequency of each word.

### --- Use only emails from the sent folders...
###     -find a way of getting all the folders while eliminating redundancy
###     -also find a way to get just the initial text
### --- Remove all punctuation.
### --- Keep track of raw words using a count vectorizer. 
### --- Group words only by individual.
### --- Eliminate words used by fewer than 10 or so individuals.
### --- Consider Ways to get rid of MetaData 
###    -(i.e. email lists, signitures, etc...)

#load email...
with open("/home/allan/Desktop/enron_fraud_detection/data/emails_text.pkl","r") as f:
    (from_emails_list,Null,from_people_list,Null)=pickle.load(f)
```

```python
qualitative_email_words = {}

for person in list(set(from_people_list)):
    qualitative_email_words[person] = ''
for i in range(len(from_people_list)):
    qualitative_email_words[from_people_list[i]] += qualitative_filter(from_emails_list[i])

qualitative_people_list = qualitative_email_words.keys()
qualitative_words = qualitative_email_words.values()
```

```python
vectorizer=TfidfVectorizer(stop_words="english")
vectorizer.fit(qualitative_email_words.values())
qualitative_words=vectorizer.get_feature_names()
qualitative_features = vectorizer.transform(qualitative_email_words.values())
```

```python
max_usage = 30
min_usage = 10

usage = sum(qualitative_features.toarray()>0)
keep_index = (np.array(usage < max_usage) & np.array(usage > min_usage) & np.array(np.array(qualitative_words)>'a'))

keep_features = qualitative_features[:,keep_index]

keep_words = np.array(qualitative_words)[keep_index.nonzero()[0]]

qualitative = {"words":keep_words}
for i in range(len(qualitative_people_list)):
    qualitative[qualitative_people_list[i]] = keep_features[i]
with open("/home/allan/Desktop/enron_fraud_detection/text_operations/qualitative.pkl","w") as f:
    pickle.dump(qualitative,f)
```
